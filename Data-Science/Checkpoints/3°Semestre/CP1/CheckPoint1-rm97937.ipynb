```python
#Importando as bibliotecas necessárias para manipular o dataset e elaborar o modelo de classificação.
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
#Fazendo a leitura do arquivo CSV contendo os dados a serem trabalhados.
dataset = pd.read_csv('onlinefoods.csv')
```

```python
#Coletando os valores categóricos para transformá-los em Dummies.
#Isso é necessário pois a biblioteca Sklearn não trabalha com valores categóricos, pois esta biblioteca se utiliza de Machine Learning para trabalhar, realizando diversas operações matemáticas que só funcionam com números.
base_categoricos = dataset[['Gender', 'Marital Status', 'Occupation', 'Monthly Income', 'Educational Qualifications', 'Output']]
base_dummies = pd.get_dummies(base_categoricos, dtype=int)
```

```python
#Separando os valores numéricos do dataset para podermos concatená-lo com a base de valores dummies.
#Fazemos isso simplesmente removendo as colunas que não devem pertencer à esta base numérica.
base_numericos = dataset.drop(['Gender', 'Marital Status', 'Occupation', 'Monthly Income', 'Educational Qualifications', 'Output', 'Feedback'], axis=1)
```

```python
#Criando a base que será usada como modelo a partir da concatenação de base_categoricos e base_numericos.
base_modelo = pd.concat([base_dummies, base_numericos], axis=1)
```

```python
#Definindo a variável dependente a partir do dataset original (que é a que será prevista).
#A variável dependente é aquela que não consegue, por si mesma, definir seu próprio valor, dependendo de outras variáveis para que ele seja determinado.
y = dataset['Feedback']
```

```python
#Definindo as bases de treino e de teste para as variáveis independentes (x) e dependentes (y).
#Precisamos fazer esta definição pois o Sklearn utilizará ambas as bases para encontrar padrões e conseguir realizar a previsão de dados.
#A base de treino vai treinar o modelo realizando diversas operações, em busca de padrões.
#A base de teste servirá para verificar a precisão da previsão e do modelo criado a partir do treinamento.
x_train, x_test, y_train, y_test = train_test_split(base_modelo, y, test_size=0.2, random_state=42)
```

```python
#Instanciando a classe DecisionTreeClassifier, que permitirá realizar tanto o treino quanto o teste do modelo.
#Devemos usar a DecisionTreeClassifier sempre que queremos prever valores categóricos, pois esta classe é feita exatamente para trabalhar com este tipo de valor.
clf = DecisionTreeClassifier()
```

```python
#Realizando o treinamento do modelo com o método fit() para podermos encontrar padrões no conjunto de dados e realizar a previsão de futuros valores.
clf.fit(x_train, y_train)
```

```python
#Realizando teste para determinar a precisão do modelo.
clf.score(x_test, y_test)
```
