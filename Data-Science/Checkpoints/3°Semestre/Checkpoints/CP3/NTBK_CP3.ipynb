```python
# prompt: Calcule a soma dos valores da coluna total por cada valor Ãºnico da coluna uf de base2021 que seja relacionado apenas ao valor COVID da coluna tipo_doenca. Exiba um grÃ¡fico sobre isso

import pandas as pd
import matplotlib.pyplot as plt

base2021 = pd.read_csv("obitos-2021.csv")

covid_deaths = base2021[base2021['tipo_doenca'] == 'COVID']
total_deaths_by_state = covid_deaths.groupby('uf')['total'].sum()

plt.figure(figsize=(10, 6))
plt.bar(total_deaths_by_state.index, total_deaths_by_state.values)
plt.xlabel('Estado')
plt.ylabel('Quantidade de Ã²bitos')
plt.title('Total de Ã³bitos por COVID por estado (2021)')
plt.show()
```

```python
# prompt: Calcule a soma dos valores da coluna total por cada valor Ãºnico da coluna uf de base2021 que seja relacionado apenas ao valor COVID da coluna tipo_doenca. Depois aplique as mÃ©tricas bÃ¡sicas da estatÃ­stica descritiva. Utilize o coeficiente de variaÃ§Ã£o ao invÃ©s da variÃ¢ncia, se vocÃª for utilizar

import pandas as pd

# Selecionar dados de COVID-19 de 2021
covid_deaths_2021 = base2021[base2021['tipo_doenca'] == 'COVID']

# Agrupar por estado e somar o total de Ã³bitos
total_deaths_by_state = covid_deaths_2021.groupby('uf')['total'].sum()

# Calcular mÃ©tricas bÃ¡sicas de estatÃ­stica descritiva
mean = total_deaths_by_state.mean().round(2)
median = total_deaths_by_state.median().round(2)
mode = total_deaths_by_state.mode().round(2)
coefficient_of_variation = (total_deaths_by_state.std() / total_deaths_by_state.mean() *100).round(2)

# Imprimir resultados
print("Medidas de tendÃªncia central:")
print(f"MÃ©dia: {mean}")
print(f"Mediana: {median}")
print(f"Moda: {mode}")

print("\nMedidas de dispersÃ£o:")
print(f"Coeficiente de variaÃ§Ã£o: {coefficient_of_variation}")

```

```python
# prompt:  Calcule a soma dos valores da coluna total por cada valor Ãºnico da coluna local_obito que seja relacionado apenas ao valor COVID da coluna tipo_doenca. Depois aplique as mÃ©tricas bÃ¡sicas da estatÃ­stica descritiva. Utilize o coeficiente de variaÃ§Ã£o ao invÃ©s da variÃ¢ncia, se vocÃª for utilizar

# Calcule a soma dos valores da coluna total por cada valor Ãºnico da coluna local_obito que seja relacionado apenas ao valor COVID da coluna tipo_doenca
soma_por_local_obito_covid = base2021[base2021['tipo_doenca'] == 'COVID'].groupby('local_obito')['total'].sum()

# Aplique as mÃ©tricas bÃ¡sicas da estatÃ­stica descritiva
media = soma_por_local_obito_covid.mean().round(2)
mediana = soma_por_local_obito_covid.median().round(2)
moda = soma_por_local_obito_covid.mode().round(2)
desvio_padrao = soma_por_local_obito_covid.std().round(2)
amplitude = soma_por_local_obito_covid.max() - soma_por_local_obito_covid.min()
assimetria = soma_por_local_obito_covid.skew().round(2)
curtose = soma_por_local_obito_covid.kurtosis().round(2)
coef_variacao = (desvio_padrao / media) * 100

# Imprima os resultados
print("Medidas de tendÃªncia central:")
print(f"MÃ©dia: {media}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda}")

print("\nMedidas de dispersÃ£o:")
print(f"Desvio padrÃ£o: {desvio_padrao}")
print(f"Amplitude: {amplitude}")
print(f"Coeficiente de variaÃ§Ã£o: {coef_variacao}%")

print("\nMedidas de forma:")
print(f"Assimetria: {assimetria}")
print(f"Curtose: {curtose}")
```
