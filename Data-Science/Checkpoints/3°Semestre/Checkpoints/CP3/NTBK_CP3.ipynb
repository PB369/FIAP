```python
# prompt: Calcule a soma dos valores da coluna total por cada valor único da coluna uf de base2021 que seja relacionado apenas ao valor COVID da coluna tipo_doenca. Exiba um gráfico sobre isso

import pandas as pd
import matplotlib.pyplot as plt

base2021 = pd.read_csv("obitos-2021.csv")

covid_deaths = base2021[base2021['tipo_doenca'] == 'COVID']
total_deaths_by_state = covid_deaths.groupby('uf')['total'].sum()

plt.figure(figsize=(10, 6))
plt.bar(total_deaths_by_state.index, total_deaths_by_state.values)
plt.xlabel('Estado')
plt.ylabel('Quantidade de òbitos')
plt.title('Total de óbitos por COVID por estado (2021)')
plt.show()
```

```python
# prompt: Calcule a soma dos valores da coluna total por cada valor único da coluna uf de base2021 que seja relacionado apenas ao valor COVID da coluna tipo_doenca. Depois aplique as métricas básicas da estatística descritiva. Utilize o coeficiente de variação ao invés da variância, se você for utilizar

import pandas as pd

# Selecionar dados de COVID-19 de 2021
covid_deaths_2021 = base2021[base2021['tipo_doenca'] == 'COVID']

# Agrupar por estado e somar o total de óbitos
total_deaths_by_state = covid_deaths_2021.groupby('uf')['total'].sum()

# Calcular métricas básicas de estatística descritiva
mean = total_deaths_by_state.mean().round(2)
median = total_deaths_by_state.median().round(2)
mode = total_deaths_by_state.mode().round(2)
coefficient_of_variation = (total_deaths_by_state.std() / total_deaths_by_state.mean() *100).round(2)

# Imprimir resultados
print("Medidas de tendência central:")
print(f"Média: {mean}")
print(f"Mediana: {median}")
print(f"Moda: {mode}")

print("\nMedidas de dispersão:")
print(f"Coeficiente de variação: {coefficient_of_variation}")

```

```python
# prompt:  Calcule a soma dos valores da coluna total por cada valor único da coluna local_obito que seja relacionado apenas ao valor COVID da coluna tipo_doenca. Depois aplique as métricas básicas da estatística descritiva. Utilize o coeficiente de variação ao invés da variância, se você for utilizar

# Calcule a soma dos valores da coluna total por cada valor único da coluna local_obito que seja relacionado apenas ao valor COVID da coluna tipo_doenca
soma_por_local_obito_covid = base2021[base2021['tipo_doenca'] == 'COVID'].groupby('local_obito')['total'].sum()

# Aplique as métricas básicas da estatística descritiva
media = soma_por_local_obito_covid.mean().round(2)
mediana = soma_por_local_obito_covid.median().round(2)
moda = soma_por_local_obito_covid.mode().round(2)
desvio_padrao = soma_por_local_obito_covid.std().round(2)
amplitude = soma_por_local_obito_covid.max() - soma_por_local_obito_covid.min()
assimetria = soma_por_local_obito_covid.skew().round(2)
curtose = soma_por_local_obito_covid.kurtosis().round(2)
coef_variacao = (desvio_padrao / media) * 100

# Imprima os resultados
print("Medidas de tendência central:")
print(f"Média: {media}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda}")

print("\nMedidas de dispersão:")
print(f"Desvio padrão: {desvio_padrao}")
print(f"Amplitude: {amplitude}")
print(f"Coeficiente de variação: {coef_variacao}%")

print("\nMedidas de forma:")
print(f"Assimetria: {assimetria}")
print(f"Curtose: {curtose}")
```
